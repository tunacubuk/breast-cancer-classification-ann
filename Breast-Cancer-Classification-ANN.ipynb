```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras  import optimizers
from tensorflow.keras.callbacks import History
```

```python
data= pd.read_csv('breast.csv', sep=',')
data.replace('?', 0, inplace=True)
data["x6"] = pd.to_numeric(data["x6"])
```

```python
dataX=data.iloc[:,1 :-1]   # baЕҹtan son sГјtuna kadar al
dataY=data.iloc[:, -1]
# Son sГјtunu al
```

```python
dataY= np.where(dataY>2, 1, 0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(dataX,dataY,test_size=0.30, random_state=42)
```

```python
from  sklearn.preprocessing  import MinMaxScaler, StandardScaler
msc = MinMaxScaler(feature_range=(0,1))
X_train = msc.fit_transform(X_train)
X_test  = msc.fit_transform(X_test)
```

```python
opt = optimizers.SGD(learning_rate=0.001,  momentum=0.6) # Г¶Дҹrenme algoritmasДұ
```

```python
model = Sequential()  # YSA aДҹДұ kuurlumu iiЕҹelmi iГ§in
history = History()
```

```python
model.add(Dense(5,  activation = 'relu' , input_dim = X_train.shape[1])) # gizli katman dГјДҹГјm 5 sayДұsДұ
model.add(Dense(1,  activation = 'sigmoid'))  # 1 Г§ДұkДұЕҹ dГјДҹГјmГј
```

```python
from keras import losses
loss_function = losses.SparseCategoricalCrossentropy(from_logits=True)
loss_function = losses.mean_squared_error;
```

```python
model.compile(optimizer = 'adam', loss='mse',  metrics = ['acc','mse'])
```

```python
print(model.summary()) # modelin parametreleri
```

```python
history= model.fit(X_train, y_train, batch_size=8, validation_split= 0.1, epochs=500)
```

```python
tahmin= model.predict(X_test)
```

```python
tahmin= np.where(tahmin>0.5, 1, 0)
```

```python
from sklearn.metrics import accuracy_score
print('ACC', round(accuracy_score(y_test, tahmin),2) ) # ACC: 0.97 olarak elde edildi. Bu iyi bir oran
```

```python
from sklearn.metrics import r2_score
print('R2', round(r2_score(y_test, tahmin),2) )   # R2: 0.87  olarak elde edildi. Д°yi bir performans olduДҹunu ifade eder
```
